# HASHING

Pencarian akan lebih cepat lagi jika semua data pada list terletak tepat berada ditempatnya masing-masing, sehingga pencarian dilakukan hanya dengan satu kali proses perbandingan saja.

### Terdapat beberapa istilah dasar sebagai berikut:

1. Hash Table, yaitu sebuah tempat penyimpanan data, yang dibuat sedemikian rupa, sehingga
dapat memudahkan pencarian.
2. Tipe data list di python dapat digunakan untuk merepresentasikan hash table slot, yaitu posisi (indeks) yang terdapat pada hash table sebagai tempat penyimpanan setiap data. Karena slot berfungsi seperti halnya indeks, maka nilai slot adalah nilai integer mulai dari nol sampai dengan ukuran dari hash table, misalkan slot 0, slot 1, slot 2, .... , slot .
3. Hash function, yaitu suatu fungsi yang memetakan antara data dengan slot di dalam hash
table.

### Berikut adalah fungsi-fungsi yang diperlukan untuk algoritma hashing ini, antara lain :

1. hash function (gunakan remainder function, yaitu data dimodulus dengan 11), nilai balik
merupakan nilai hash
2. createHashTable, untuk membuat hash table, dengan inisialisasi semua slot berisi 'none'
3. putData, yaitu menyusun data ke dalam hash table, berdasarkan nilai hash yang dihasilkan
4. searchHash, argumen merupakan data yang dicari, dan nilai balik berupa True or False, yaitu
apakah data ditemukan di dalam hash table.

## Contoh implementasi hashing

```python
def remainderFunction (data, num):
    return (data%num)

def createHashTable(num):
    temp = []
    for i in range(num):
        temp.append('none')
    return (temp)
def putData(data,table):
    for i in range(len(data)):
        ind = remainderFunction(data[i],len(table))
        table[ind] = data [i]
    return(table)
def searchHash(data,table):
    hasVal = remainderFunction(data,len(table))
    if data == table[hasVal]:
        return True
    else:
        return False

a = [54, 93, 17, 77, 31]
hashTable = createHashTable(11)
print(hashTable)

hashTable = putData(a,hashTable)
print(hashTable)

searchHash(93,hashTable)
```

```python
class Hash:
    def __init__(self, buckets):
        # Number of buckets
        self.bucket_count = buckets

        # Create empty chains for each bucket
        self.table = [[] for _ in range(self.bucket_count)]

    # Insert a key into hash table
    def insert(self, key):
        # Compute hash index
        index = self.get_hash_index(key)

        # Append key to corresponding chain
        self.table[index].append(key)

    # Remove a key from hash table
    def remove(self, key):
        # Compute hash index
        index = self.get_hash_index(key)

        # Remove key if present in chain
        if key in self.table[index]:
            self.table[index].remove(key)

    # Display contents of hash table
    def display(self):
        for i in range(self.bucket_count):
            print(i, end="")

            # Print all keys in chain
            for key in self.table[i]:
                print(" -->", key, end="")

            print()

    # Hash function
    def get_hash_index(self, key):
        return key % self.bucket_count

if __name__ == "__main__":
    keys = [7, 18, 12, 25]

    hash_table = Hash(7)

    for key in keys:
        hash_table.insert(key)

    hash_table.remove(12)
    hash_table.display()
```

## Penanganan Collusin
Penanganan collusion ini dapat dilakukan dengan cara mencari kembali di dalam hash table slot yang kosong untuk data yang menyebabkan terjadinya collusion. Penanganan ini dikenal dengan konsep open addressing, yaitu pencarian slot kosong yang siap ditempati oleh data baru.

# Open Adress
## 1. Linear Probing
Linear Probing yaitu mencari slot kosong dengan cara mengunjungi slot-slot ini satu persatu, dimulai dari slot tempat terjadinya collusion.

## cara kerja nya 
Jika suatu index sudah terisi, maka:<br>
1. sistem mencari slot kosong berikutnya,
2. dengan maju satu per satu secara linear.

## rumus linear probing
(h(k)+i)mod m<br>
Keterangan:<br>
- h(k) = hash awal
- i = langkah perpindahan
- m = ukuran tabel

## Contoh 
ukuran tabel = 10 <br>
memasukkan data 25 <br>
25 mod 10 = 5 (disimpan di indeks 5)<br>
memasukkan data 35<br>
35 mod 10 = 5<br>
indeks 5 sudah terisi maka terjadi collision<br>
cek indeks berikutnya dan simpan 35 di index 6 <br>
dan seterusnya 

## Contoh implementasi Open Adress pada Linear Probing

```python
size = 10
hash_table = [None] * size

def hash_function(data):
    return data % size

def insert(data):
    index = hash_function(data)
    # jika index sudah terisi
    while hash_table[index] is not None:
        # pindah ke index berikutnya
        index = (index + 1) % size
    hash_table[index] = data

insert(25)
insert(35)
insert(45)
insert(15)
print(hash_table)
```

## 2. Quadratic Probing
Quadratic Probing adalah metode untuk mengatasi collision pada hashing dengan cara:<br>
mencari slot kosong menggunakan lompatan kuadrat.

## Rumus Quadratic Probing

(h(k)+i^2)mod m<br>

Keterangan:<br>
- h(k) = hash awal
- i = percobaan
- m = ukuran tabel

## Contoh
ukuran tabel = 10<br>
memasukkan data 25<br>
25 mod 10 = 5 (disimpan di indeks 5)<br>
memasukkan data 35<br>
35 mod 10 = 5<br>
indeks 5 sudah terisi 25 maka terjadi collision<br>
quadratic mulai di hitung<br>
5 + 1² = 6<br>
cek index 6 kosong maka 35 disimpan di index 6<br>
memasukkan data 45<br>
45 mod 10 = 5<br>
index 5 penuh , di cek lagi index 6 penuh maka hitung quadratic ke 2<br>
5 + 2² = 9<br>
index 9 kosong, maka 45 di simpan di index 9<br>
dan seterusnya

## Contoh implementasi Open Adress pada Quadratic Probing

```python
size = 10
hash_table = [None] * size

# fungsi hash
def hash_function(data):
    return data % size

# insert quadratic probing
def insert(data):
    index = hash_function(data)
    i = 1

    # jika collision
    while hash_table[index] is not None:
        index = (hash_function(data) + i**2) % size
        i += 1
    hash_table[index] = data

insert(25)
insert(35)
insert(45)
insert(15)
print(hash_table)
```

# Close Adress
## Chaining
Chaining adalah metode untuk mengatasi collision pada hashing dengan cara:<br>
menyimpan beberapa data dalam satu index menggunakan linked list / list.

## Contoh
ukuran tabel = 10<br>
memasukkan data 25<br>
25 mod 10 = 5 (disimpan di indeks 5)<br>
memasukkan data 35<br>
35 mod 10 = 5<br>
indeks 5 sudah terisi 25, karena chaining 35 ditambahkan di belakang<br>
5 → 25 → 35<br>
memasukkan data 45<br>
5 → 25 → 35 → 45 

## Contoh implmentasi Close Adress pada Chaining

```python
size = 10
# membuat hash table berupa list kosong
hash_table = [[] for _ in range(size)]

# fungsi hash
def hash_function(data):
    return data % size

# insert chaining
def insert(data):
    index = hash_function(data)
    hash_table[index].append(data)
    
# insert data
insert(25)
insert(35)
insert(45)
print(hash_table)
```

## Mid-Square Method (Fungsi Hash)

Mid-Square Method adalah metode hashing dengan cara:<br>
mengkuadratkan data terlebih dahulu, lalu mengambil beberapa digit tengah sebagai index hash.

## Cara Kerja Mid-Square Method
- Ambil data (key)
- Kuadratkan data
- Ambil angka di tengah
- Gunakan sebagai index hash

## Contoh implementasi Mid-Square Method

```python
def midFunction(data, table_size):
    square = str(data ** 2)

    midInd = len(square) // 2

    if len(square) < 3:
        midVal = int(square)

    elif len(square) % 2 == 0:
        midVal = int(square[midInd - 1: midInd + 1])

    else:
        midVal = int(square[midInd])

    return midVal % table_size

midFunction(12, 10)
```
