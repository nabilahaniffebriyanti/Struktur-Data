# HASHING

### 1. Close Address (Chaining)--> CRU(cari diganti dengan yang baru)D(cari di delete)

```python
table = [[] for i in range(11)]

# Fungsi hash
def hash_func(x):
    return x % 11

# CREATE / INSERT
def insert(table, key):
    table[hash_func(key)].append(key)

# READ / SEARCH
def search(table, key):
    index = hash_func(key)
    return key in table[index]

# DELETE
def delete(table, key):
    index = hash_func(key)
    if key in table[index]:
        table[index].remove(key)
        return True
    return False

# UPDATE (versi perbaikan)
def update(table, old_key, new_key):
    if delete(table, old_key):
        insert(table, new_key)
        return True
    return False

# Data awal
data = [54, 26, 93, 17, 77, 31, 44, 55, 20]

print("Awal:", table)

# Insert data
for d in data:
    insert(table, d)

print("Setelah Insert:", table)

# SEARCH
print("\n--- SEARCH ---")
print("Cari 77:", search(table, 77))
print("Cari 100:", search(table, 100))

# UPDATE
print("\n--- UPDATE ---")
print("Update 77 -> 99:", update(table, 77, 99))
print("Hasil:", table)

# DELETE
print("\n--- DELETE ---")
print("Hapus 31:", delete(table, 31))
print("Hasil:", table)
```

### - CRUD --> Linear Probbing --> Fungsi hash biasa

```python
table = [None] * 11
DELETED = "<deleted>"

# Fungsi hash biasa (modulo)
def hash_func(x):
    return x % 11

# CREATE / INSERT
def insert(table, key):
    index = hash_func(key)

    while table[index] is not None and table[index] != DELETED:
        index = (index + 1) % 11   # Linear Probing

    table[index] = key

# READ / SEARCH
def search(table, key):
    index = hash_func(key)
    start = index

    while table[index] is not None:
        if table[index] == key:
            return True

        index = (index + 1) % 11

        if index == start:
            break

    return False

# DELETE
def delete(table, key):
    index = hash_func(key)
    start = index

    while table[index] is not None:
        if table[index] == key:
            table[index] = DELETED
            return True

        index = (index + 1) % 11

        if index == start:
            break

    return False

# UPDATE
def update(table, old_key, new_key):
    if delete(table, old_key):
        insert(table, new_key)
        return True
    return False

# LOAD FACTOR
def load_factor(table):
    filled = 0

    for item in table:
        if item is not None and item != DELETED:
            filled += 1

    return filled / len(table)

# Data awal
data = [54, 26, 93, 17, 77, 31, 44]

# INSERT
for d in data:
    insert(table, d)

print("Setelah Insert:")
print(table)

# LOAD FACTOR
lf = load_factor(table)

print("\nLoad Factor =", lf)

if lf < 1:
    print("Load Factor < 1 (Valid)")
else:
    print("Load Factor >= 1 (Table Full)")

# SEARCH
print("\n--- SEARCH ---")
print("Cari 77:", search(table, 77))
print("Cari 100:", search(table, 100))

# UPDATE
print("\n--- UPDATE ---")
print("Update 77 -> 99:", update(table, 77, 99))
print("Hasil:", table)

# DELETE
print("\n--- DELETE ---")
print("Hapus 26:", delete(table, 26))
print("Hasil:", table)
```

### - CRUD --> Quadratic Probbing --> Fungsi hash biasa

```python
table = [None] * 11
DELETED = "<deleted>"

# Fungsi hash biasa (modulo)
def hash_func(x):
    return x % 11

# LOAD FACTOR
def load_factor(table):
    filled = 0

    for item in table:
        if item is not None and item != DELETED:
            filled += 1

    return filled / len(table)

# CREATE / INSERT
def insert(table, key):

    lf = load_factor(table)

    if lf >= 1:
        print("Table Full")
        return

    index = hash_func(key)
    i = 1

    while table[index] is not None and table[index] != DELETED:
        index = (hash_func(key) + i**2) % 11   # Quadratic Probing
        i += 1

    table[index] = key

# READ / SEARCH
def search(table, key):
    index = hash_func(key)
    start = index
    i = 1

    while table[index] is not None:

        if table[index] == key:
            return True

        index = (hash_func(key) + i**2) % 11
        i += 1

        if index == start:
            break

    return False

# DELETE
def delete(table, key):
    index = hash_func(key)
    start = index
    i = 1

    while table[index] is not None:

        if table[index] == key:
            table[index] = DELETED
            return True

        index = (hash_func(key) + i**2) % 11
        i += 1

        if index == start:
            break

    return False

# UPDATE
def update(table, old_key, new_key):
    if delete(table, old_key):
        insert(table, new_key)
        return True
    return False

# Data awal
data = [54, 26, 93, 17, 77, 31, 44]

# INSERT
for d in data:
    insert(table, d)

print("Setelah Insert:")
print(table)

# LOAD FACTOR
lf = load_factor(table)

print("\nLoad Factor =", lf)

if lf < 1:
    print("Load Factor < 1 (Valid)")
else:
    print("Load Factor >= 1 (Table Full)")

# SEARCH
print("\n--- SEARCH ---")
print("Cari 77:", search(table, 77))
print("Cari 100:", search(table, 100))

# UPDATE
print("\n--- UPDATE ---")
print("Update 77 -> 99:", update(table, 77, 99))
print("Hasil:", table)

# DELETE
print("\n--- DELETE ---")
print("Hapus 26:", delete(table, 26))
print("Hasil:", table)
```

### Contoh Create (Insert) dengan Load Factor < 1
### Menggunakan Linear Probbing

```python
table = [None] * 10
DELETED = "<deleted>"

# Fungsi hash biasa
def hash_func(x):
    return x % 10

# Hitung Load Factor
def load_factor(table):
    filled = 0

    for item in table:
        if item is not None and item != DELETED:
            filled += 1

    return filled / len(table)

# INSERT Linear Probing
def insert(table, key):

    lf = load_factor(table)

    # Cek apakah tabel penuh
    if lf >= 1:
        print("Table Full, tidak bisa insert", key)
        return

    index = hash_func(key)

    while table[index] is not None and table[index] != DELETED:
        index = (index + 1) % 10

    table[index] = key

    print(f"Insert {key} berhasil")
    print("Load Factor =", load_factor(table))
    print(table)
    print()

# Data hampir penuh
data = [11, 22, 33, 44, 55, 66, 77, 88, 99]

# INSERT DATA
for d in data:
    insert(table, d)

# Tambah 1 data lagi
insert(table, 100)

# Tambah lagi sampai penuh
insert(table, 111)

print("Hash Table Akhir:")
print(table)
```

# 3. MidSquareMethod

### CRUD --> Linear Probbing --> Fungsi hash Mid Square Method

```python
table = [None] * 11
DELETED = "<deleted>"

# Fungsi Hash Mid Square Method
def midsquare_hash(x):
    square = x * x
    square_str = str(square)

    mid = len(square_str) // 2

    # Ambil digit tengah
    middle_digit = int(square_str[mid])

    return middle_digit % 11

# LOAD FACTOR
def load_factor(table):
    filled = 0

    for item in table:
        if item is not None and item != DELETED:
            filled += 1

    return filled / len(table)

# CREATE / INSERT
def insert(table, key):

    lf = load_factor(table)

    if lf >= 1:
        print("Table Full")
        return

    index = midsquare_hash(key)

    while table[index] is not None and table[index] != DELETED:
        index = (index + 1) % 11   # Linear Probing

    table[index] = key

# READ / SEARCH
def search(table, key):
    index = midsquare_hash(key)
    start = index

    while table[index] is not None:
        if table[index] == key:
            return True

        index = (index + 1) % 11

        if index == start:
            break

    return False

# DELETE
def delete(table, key):
    index = midsquare_hash(key)
    start = index

    while table[index] is not None:
        if table[index] == key:
            table[index] = DELETED
            return True

        index = (index + 1) % 11

        if index == start:
            break

    return False

# UPDATE
def update(table, old_key, new_key):
    if delete(table, old_key):
        insert(table, new_key)
        return True
    return False

# Data awal
data = [54, 26, 93, 17, 77, 31, 44]

# INSERT
for d in data:
    insert(table, d)

print("Setelah Insert:")
print(table)

# LOAD FACTOR
lf = load_factor(table)

print("\nLoad Factor =", lf)

if lf < 1:
    print("Load Factor < 1 (Valid)")
else:
    print("Load Factor >= 1 (Table Full)")

# SEARCH
print("\n--- SEARCH ---")
print("Cari 77:", search(table, 77))
print("Cari 100:", search(table, 100))

# UPDATE
print("\n--- UPDATE ---")
print("Update 77 -> 99:", update(table, 77, 99))
print("Hasil:", table)

# DELETE
print("\n--- DELETE ---")
print("Hapus 26:", delete(table, 26))
print("Hasil:", table)
```
