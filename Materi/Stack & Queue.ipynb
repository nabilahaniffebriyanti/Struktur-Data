# STACK
Stack adalah tumpukan data yang mengikuti aturan LIFO (Last In, First Out), yang berarti elemen terakhir yang dimasukkan ke dalam stack akan menjadi elemen pertama yang dihapus dari stack. (bisa di akses secara tumpukan atau antrian).

## Ada 5 operasi yang ada di Stack
1. PUSH<br>
   penambahan data baru pada posisi top dari stack.<br>
3. POP<br>
   penghapusan data yang terdapat di posisi top dari stack.<br>
5. PEEK<br>
   informasi data yang terletak pada posisi top.<br>
7. IsEmpty<br>
   untuk mengecek apakah stack tersebut kosong atau tidak.<br>
9. SIZE<br>
   untuk mengetahui jumlah data yang ada dalam stack.<br>

```python
stack = []

#PUSH
#penambahan data baru pada posisi top dari stack.
stack.append('B')
stack.append('I')
stack.append('L')
stack.append('A')
print("stack: ", stack)

#POP
#penghapusan data yang terdapat di posisi top dari stack.
pop = stack.pop()
print("Pop:", pop)

#PEEK
#informasi data yang terletak pada posisi top.
peek = stack[-1]
print("Peek: ", peek)

#IsEmpty
#untuk mengecek apakah stack tersebut kosong atau tidak.
isEmpty = not (stack)
print("isEmpty: ", isEmpty)

#SIZE
#untuk mengetahui jumlah data yang ada dalam stack.
print("Size: ",len(stack))
```

```python
class Stack:
    data = []

    def push(self, huruf):
        self.data.append(huruf)

    def pop(self):
        return self.data.pop()

    def size(self):
        return len(self.data)

kata_asli = "KATAK"
s = Stack()

for h in kata_asli:
    s.push(h)

kata_terbalik = ""
while s.size() > 0:
    kata_terbalik += s.pop()

print("Kata Asli:", kata_asli)
print("Kata Terbalik:", kata_terbalik)

if kata_asli == kata_terbalik:
    print("KESIMPULAN: Palindrome")
else:
    print("KESIMPULAN: Bukan Palindrome")
```

## CONTOH LATIHAN
Riwayat Undo (Simulasi Undo Text Editor)
Buatlah program sederhana untuk mensimulasikan fitur Undo pada editor teks.<br>
Ketentuan:<br>
User memasukkan beberapa teks.<br>
Setiap input disimpan ke dalam stack.<br>
Jika user mengetik UNDO, maka teks terakhir dihapus.<br>

Contoh:<br>
Input: Halo<br>
Input: Dunia<br>
Input: Belajar Stack<br>
Input: UNDO<br>
Output: Dunia<br>

```python
# Membuat stack kosong
stack = []

while True:
    teks = input("Input: ")

    # Jika user mengetik EXIT, program berhenti
    if teks == "EXIT":
        break

    # Jika user mengetik UNDO
    elif teks == "UNDO":
        if len(stack) > 0:
            stack.pop()   # Menghapus teks terakhir
        else:
            print("Stack kosong!")

    # Menyimpan input ke stack
    else:
        stack.append(teks)

# Menampilkan isi terakhir stack
if len(stack) > 0:
    print("Output:", stack[-1])
else:
    print("Tidak ada teks")
```

## CONTOH SOAL

[10, 20, 30, 40, 50]<br>
Buat program untuk menghapus angka 30 tanpa merusak urutan elemen lainnya.<br>

```python
stack = [10, 20, 30, 40, 50]
temp = []

# pindahkan elemen sampai ketemu 30
while stack:
    x = stack.pop()
    if x == 30:
        break
    temp.append(x)

# kembalikan elemen lain tanpa 30
while temp:
    stack.append(temp.pop())

print("Stack setelah hapus 30:", stack)
```

# Queue
Queues atau antrian merupakan struktur data dimana penambahan data baru dan penghapusan data berada di ujung yang berbeda.<br>
pada Queues menggunakan konsep FIFO (First In First Out).konsep di mana elemen yang pertama masuk akan keluar pertama.<br>
Pada Queues, seperti halnya antrian, penambahan data baru dilakukan di suatu ujung atau yang dikenal dengan nama rear, dan penghapusan data dilakukan pada ujung yang dikenal dengan nama front.

# Operasi pada Queue

- Queue() membuat queue baru yang kosong. Tidak memerlukan parameter dan mengembalikan queue kosong.<br>
- enqueue(item) menambahkan item baru ke bagian belakang queue. Memerlukan item dan tidak mengembalikan apa pun.<br>
- dequeue() menghapus item paling depan dari queue. Tidak memerlukan parameter dan mengembalikan item tersebut. Queue akan berubah.<br>
- isEmpty() mengecek apakah queue kosong. Tidak memerlukan parameter dan mengembalikan nilai boolean.<br>
- size() mengembalikan jumlah item dalam queue. Tidak memerlukan parameter dan mengembalikan bilangan integer.<br>

```python
queue = []

# enqueue (tambah data)
def enqueue(data):
    queue.append(data)

# dequeue (ambil data)
def dequeue():
    if len(queue) == 0:
        return "Queue kosong"
    return queue.pop(0)

# contoh penggunaan
enqueue(10)
enqueue(20)
enqueue(30)

print("Queue:", queue)

print("Dequeue:", dequeue())
print("Queue setelah dequeue:", queue)
```

# Dequeue
Juga dikenal sebagai double-ended queue, adalah koleksi item yang terurut yang mirip dengan queue. Memiliki dua ujung, yaitu front dan rear, dan item-item tetap berada pada posisinya di dalam koleksi.<br>
Operasi dalam struktur data queue (antrian) yang berarti mengambil atau menghapus elemen dari antrian.<br>
Dalam konsep queue (FIFO = First In, First Out), proses dequeue dilakukan pada bagian depan (front/head) antrian.<br>

# Operasi pada Dequeue
- Deque() membuat deque baru yang kosong. Tidak memerlukan parameter dan mengembalikan deque kosong.<br>
- addFront(item) menambahkan item baru ke bagian depan (front) deque. Memerlukan item dan tidak mengembalikan apa pun.<br>
- addRear(item) menambahkan item baru ke bagian belakang (rear) deque. Memerlukan item dan tidak mengembalikan apa pun.<br>
- removeFront() menghapus item paling depan dari deque. Tidak memerlukan parameter dan mengembalikan item tersebut. Deque akan berubah.<br>
- removeRear() menghapus item paling belakang dari deque. Tidak memerlukan parameter dan mengembalikan item tersebut. Deque akan berubah.<br>
- isEmpty() mengecek apakah deque kosong. Tidak memerlukan parameter dan mengembalikan nilai boolean.<br>
- size() mengembalikan jumlah item dalam deque. Tidak memerlukan parameter dan mengembalikan bilangan integer.<br>

```python
queue = [10, 20, 30]

# fungsi dequeue
def dequeue():
    if len(queue) == 0:
        return "Queue kosong"
    return queue.pop(0)

# contoh penggunaan
print(dequeue())  
print(queue)      
```

# Deques (double-ended queue)
Deque (singkatan dari double-ended queue) adalah struktur data yang memungkinkan penambahan dan penghapusan elemen dari kedua ujung, yaitu bagian depan (front) dan belakang (rear).<br>
Deques dapat dilakukan di kedua ujung, seperti halnya queues, hanya terdapat sedikit perbedaan.<br>
penambahan data dapat dilakukan pada kedua ujung, baik front maupun rear. Begitu juga halnya dengan penghapusan data yang dapat dilakukan pada front maupun rear.

# Operasi pada Deques
- deque (), inisialisasi struktur data deque kosong
- addFront (data), penambahan data baru di ujung front pada deque
- addRear (data), penambahan data baru di ujung rear pada deque
- removeRear (), penghapusan data pada ujung rear
- removeFront (), penghapusan data pada ujung front
- isEmpty(), pengecekan apakah deque dalam keadaan kosong
- size (), informasi jumlah data yang terdapat pada deque

```python
from collections import deque

stack = deque()

stack.append(1)
stack.append(2)
stack.append(3)

print(stack.pop())  # 3
print(stack.pop())  # 2
```

```python
from collections import deque

d = deque([1, 2, 3, 4])

d.pop()        # hapus dari belakang (4)
d.popleft()    # hapus dari depan (1)

print(d)  # deque([2, 3])
```

```python
class Queue:
   def __init__(self):
       self.items = []
 
   def is_empty(self):
       return len(self.items) == 0
 
   def enqueue(self, item):
       self.items.append(item)
 
   def dequeue(self):
       if not self.is_empty():
           return self.items.pop(0)
       else:
           return "Queue is empty"
 
   def size(self):
       return len(self.items)
 
# Contoh penggunaan
antrian = Queue()
 
# Menambahkan elemen ke dalam antrian
antrian.enqueue("Pelanggan 1")
antrian.enqueue("Pelanggan 2")
antrian.enqueue("Pelanggan 3")
 
# Melihat ukuran antrian
print("Ukuran antrian:", antrian.size())
 
# Mengambil elemen dari antrian
print("Panggilan ke kasir:", antrian.dequeue())
 
# Melihat ukuran antrian setelah panggilan pertama
print("Ukuran antrian setelah panggilan pertama:", antrian.size())
```

```python
class Deque:
    def __init__(self):
        self.items = []

    # Menambahkan elemen di depan
    def addFront(self, item):
        self.items.insert(0, item)

    # Menambahkan elemen di belakang
    def addRear(self, item):
        self.items.append(item)

    # Menghapus elemen dari depan
    def removeFront(self):
        if not self.isEmpty():
            return self.items.pop(0)
        else:
            return "Deque kosong"

    # Menghapus elemen dari belakang
    def removeRear(self):
        if not self.isEmpty():
            return self.items.pop()
        else:
            return "Deque kosong"

    # Cek apakah kosong
    def isEmpty(self):
        return len(self.items) == 0

    # Menampilkan isi deque
    def display(self):
        return self.items


# Contoh penggunaan
dq = Deque()

dq.addRear(10)
dq.addRear(20)
dq.addFront(5)
dq.addFront(1)

print("Isi deque:", dq.display())

dq.removeFront()
print("Setelah removeFront:", dq.display())

dq.removeRear()
print("Setelah removeRear:", dq.display())
```

```python
from collections import deque

def is_palindrome(kata):
    d = deque(kata)
    
    while len(d) > 1:
        if d.popleft() != d.pop():
            return False
    return True

# Contoh penggunaan
print(is_palindrome("katak"))   # True
print(is_palindrome("halo"))    # False
```
