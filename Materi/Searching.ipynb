# SEARCHING / pencarian
Searching adalah proses untuk mencari data tertentu di dalam sebuah kumpulan data (list/array).<br>
Bertujuan untuk Mengetahui apakah data ada atau tidak Mengetahui posisi/index data.

# 1. LINEAR SEARCH
Linear search atau Sequential Search merupakan teknik pencarian data dengan cara membandingkan data yang
dicari dengan seluruh data yang terdapat pada kumpulan data secara satu persatu, mulai dari data
pertama sampai dengan data terakhir.<br>
data nya masih mau terurut atau udah terurut<br>
dan tidak ada keharusan kalo kondisi nya terurut<br>

## Contoh implementasi linear search

```python
data = [1, 7, 11, 23, 39, 43, 56]
cari = 10
found = False

for i in range(len(data)):
    if data[i] == cari:
        found = True
        posisi = i

if found:
    print("Data ditemukan")
    print("Posisi index:", posisi)
else:
    print("Data tidak ditemukan")
```

```python
def linearSearch(listData, value):
    index = 0
    
    while (index < len(listData) and listData[index] < value):
        index += 1
        
    if (index >= len(listData) or listData[index] != value):
        return -1
        
    return index

data = [2, 4, 6, 8, 10, 12]
cari = 8
hasil = linearSearch(data, cari)

if hasil != -1:
    print("Data ditemukan di index ke-", hasil)
else:
    print("Data tidak ditemukan")
```

## Contoh implementasi Unordered List sequential search

```python
def seqSearch(listData, data):
    ind = 0
    found = False
    
    while ind < len(listData) and not found:
        if listData[ind] == data:
            found = True
        else:
            ind = ind + 1
    return found

a = [12, 5, 9, 8, 1, 10, 26]
seqSearch(a,1)
```

## Reverse Linear Search (unordered)

```python
## proses pencarian dimulai dari elemen terakhir (ujung kanan list), bukan dari awal.
search = 2
my_list = [1, 2, 3, 4, 5, 6]

position = len(my_list) - 1
last = 0
found = False

while position >= last and not found:
    if my_list[position] == search:
        found = True
    else:
        position = position - 1

if found:
    print("Found the search number.")
    print(position)
else:
    print("Did not find the search number.")
```

# 2. BINARY SEARCH 

pencarian dilakukan pertama kali di indeks list yang berada ditengah. Jika data yang dicari sama dengan data pada indeks tersebut, maka pencarian berakhir.<br>
Binary Search adalah algoritma pencarian data yang bekerja dengan cara membagi data menjadi dua bagian secara terus-menerus sampai data ditemukan atau tidak ada.

### Syarat Binary Search
- Data harus sudah terurut (ascending atau descending).
- Jika data belum terurut, maka harus di-sort terlebih dahulu.

### Cara Kerja Binary Search
1. Tentukan:
- first → indeks awal
- last → indeks akhir
2. Cari nilai tengah (middle / mid).(mid = (first + last) // 2 )
3. Bandingkan nilai tengah dengan data yang dicari:
- Jika A[mid] == key → data ditemukan.
- Jika A[mid] < key → pencarian dilanjutkan ke kanan.
first = mid + 1
- Jika A[mid] > key → pencarian dilanjutkan ke kiri.
last = mid - 1
4. Ulangi langkah sampai data ditemukan atau first > last.

## Contoh implementasi pada binary search

```python
def binarySearch(listData, data):
    first = 0
    last = len(listData) - 1
    found = False

    while first <= last and not found:
        midpoint = (first + last) // 2
        if listData[midpoint] == data:
            found = True

        else:
            if data < listData[midpoint]:
                last = midpoint - 1
            else:
                first = midpoint + 1

    return found

a = [4,6,10,34,56,78,99]
print(binarySearch(a,34))
```

## Binary Search Rekursif

```python
def binarySearch2(listData, data):
    if len(listData) == 0:
        return False
    else:
        midpoint = len(listData) // 2
        if listData[midpoint] == data:
            return True
        else:
            if data < listData[midpoint]:
                return binarySearch2(listData[:midpoint], data)
            else:
                return binarySearch2(listData[midpoint + 1:], data)

a = [4,6,10,34,56,78,99]
binarySearch2(a, 78)
```
