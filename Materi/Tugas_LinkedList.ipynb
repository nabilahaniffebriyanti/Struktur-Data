## 1. Unorder List
#### Search Cari angka tertentu (18)
#### 1) jika ada disisipkan setelah yang dicari ada <br> Head--> 7--> 18 --> 6 --> None  Head--> 7--> 18 --> 18 --> 6 --> None  
#### 2) kalau tidak disisipkan di akhir <br> Head--> 7--> 10 --> 6 --> None  Head--> 7--> 10 --> 6 --> 18 --> None <br> insert 18 --> menggabungkan 18 Link List <br> 6 sambung 18 (setnext = temp)


```python
class Node:
    def __init__(self, init_data):
        self.data = init_data
        self.next = None
    def getData(self):
        return self.data
    def getNext(self):
        return self.next
    def setData(self, newdata):
        self.data = newdata
    def setNext(self, new_next):
        self.next = new_next

class LinkedList:
    def __init__(self):
        self.head = None
    def isEmpty(self):
        return self.head == None
    def add(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp
        
    # tampilkan isi linked list
    def display(self):
        curent = self.head
        while curent != None:
            print(curent.getData(), end=" --> ")
            curent = curent.getNext()
        print("None")
        
    # search lalu insert
    def searchInsert(self, item):
        curent = self.head
        found = False
        while curent != None and not found:

            # jika ditemukan
            if curent.getData() == item:
                found = True
                temp = Node(item)
                temp.setNext(curent.getNext())
                curent.setNext(temp)
                print(item, "ditemukan, disisipkan setelah data")
                return
            else:
                curent = curent.getNext()

        # jika tidak ditemukan
        temp = Node(item)

        # jika linked list kosong
        if self.head == None:
            self.head = temp
        else:
            curent = self.head
            while curent.getNext() != None:
                curent = curent.getNext()
            curent.setNext(temp)
        print(item, "tidak ditemukan, ditambahkan di akhir")

mylist = LinkedList()

mylist.add(6)
mylist.add(18)
mylist.add(7)

print("Sebelum:")
mylist.display()

mylist.searchInsert(18)

print("Sesudah:")
mylist.display()
```

## 2. OrderList (iterasi)
#### - cari < sekarang --> lebih besar berhenti
#### 1) jika ada --> sisipkan setelah yang dicari
#### 2) tidak ada --> disisipkan sebelumnya

```python
class Node:
    def __init__(self, init_data):
        self.data = init_data
        self.next = None
    def getData(self):
        return self.data
    def getNext(self):
        return self.next
    def setData(self, newdata):
        self.data = newdata
    def setNext(self, new_next):
        self.next = new_next

class OrderedLinkedList:
    def __init__(self):
        self.head = None
    def isEmpty(self):
        return self.head == None

    # tampilkan data
    def display(self):
        curent = self.head
        while curent != None:
            print(curent.getData(), end=" --> ")
            curent = curent.getNext()
        print("None")

    # add terurut
    def add(self, item):
        curent = self.head
        previous = None
        stop = False
        while curent != None and not stop:
            if curent.getData() > item:
                stop = True
            else:
                previous = curent
                curent = curent.getNext()

        temp = Node(item)

        if previous == None:
            temp.setNext(self.head)
            self.head = temp
        else:
            temp.setNext(curent)
            previous.setNext(temp)

    # search lalu insert
    def searchInsert(self, item):

        curent = self.head
        previous = None
        found = False
        stop = False
        while curent != None and not found and not stop:

            # jika ditemukan
            if curent.getData() == item:
                found = True

                temp = Node(item)

                temp.setNext(curent.getNext())
                curent.setNext(temp)

                print(item, "ditemukan")
                return

            else:

                # berhenti jika data sekarang lebih besar
                if curent.getData() > item:
                    stop = True

                else:
                    previous = curent
                    curent = curent.getNext()

        # jika tidak ditemukan
        temp = Node(item)

        # insert di awal
        if previous == None:
            temp.setNext(self.head)
            self.head = temp

        else:
            temp.setNext(curent)
            previous.setNext(temp)

        print(item, "tidak ditemukan")

myList = OrderedLinkedList()

myList.add(3)
myList.add(7)
myList.add(10)
myList.add(15)

print("Sebelum:")
myList.display()

myList.searchInsert(12)

print("Sesudah:")
myList.display()
```
