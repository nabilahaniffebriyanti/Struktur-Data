## 1. BUBBLE SORT
Bubble adalah algoritma pertukaran/ pengurutan sederhana yang berulang kali menelusuri daftar yang akan diurutkan, membandingkan setiap pasangan item yang berdekatan dan menukarnya jika urutannya salah. yang membutuhkan waktu lama

## Step pengurutan bubble short

- Queue() membuat queue baru yang kosong. Tidak memerlukan parameter dan mengembalikan queue kosong.<br>
- enqueue(item) menambahkan item baru ke bagian belakang queue. Memerlukan item dan tidak mengembalikan apa pun.<br>
- dequeue() menghapus item paling depan dari queue. Tidak memerlukan parameter dan mengembalikan item tersebut. Queue akan berubah.<br>
- isEmpty() mengecek apakah queue kosong. Tidak memerlukan parameter dan mengembalikan nilai boolean.<br>
- size() mengembalikan jumlah item dalam queue. Tidak memerlukan parameter dan mengembalikan bilangan integer.<br>

```python
def bubbleSort(arr):
    n = len(arr)
    
    # Traverse through all array elements
    for i in range(n):
        swapped = False

        # Last i elements are already in place
        for j in range(0, n-i-1):

            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater
            # than the next element
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if (swapped == False):
            break

# Driver code to test above
if __name__ == "__main__":
    arr = [64, 34, 25, 12, 22, 11, 90]

    bubbleSort(arr)

    print("Sorted array:")
    for i in range(len(arr)):
        print("%d" % arr[i], end=" ")
```

## Contoh implementasi bubble short Ascending (terkecil ke besar)

```python
def bubble_sort(arr): #ascending
    n = len(arr)
    
    for i in range(n):
        swapped = False
        
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        
        # Jika tidak ada pertukaran, berarti sudah terurut
        if not swapped:
            break
    
    return arr

data = [64, 34, 25, 12, 22, 11, 90]
print("Sebelum:", data)

hasil = bubble_sort(data)
print("Sesudah:", hasil)
```

## Contoh implementasi bubble short Descending (terbesar ke kecil)

```python
def bubble_sort_desc(arr): #descending
    n = len(arr)
    
    for i in range(n):
        for j in range(0, n - i - 1):
            # Tukar jika elemen sekarang lebih kecil dari berikutnya
            if arr[j] < arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    
    return arr

data = [64, 34, 25, 12, 22, 11, 90]
print("Sebelum:", data)

hasil = bubble_sort_desc(data)
print("Sesudah:", hasil)
```

## 2. SELECTION SORT
Mencari nilai minimum dalam array dan memindahkannya ke posisi pertama. Langkah ini kemudian diulangi untuk nilai terendah kedua, lalu ketiga, dan seterusnya hingga array diurutkan.

## Cara Kerja Selection Sort
1. Cari nilai terkecil dalam list.
2. Tukar dengan elemen pertama.
3. Cari nilai terkecil berikutnya.
4. Tukar dengan elemen kedua.
5. Ulangi sampai data selesai diurutkan.

```python
lst = [4, 7, 1, 9, 3, 10, 6]

print("Sebelum sorting:", lst)

for i in range(0, len(lst) - 1):
    min_idx = i

    for j in range(i + 1, len(lst)):
        if lst[j] < lst[min_idx]:
            min_idx = j

    # Tukar elemen
    lst[i], lst[min_idx] = lst[min_idx], lst[i]

print("Setelah sorting:", lst)
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
      
        # Assume the current position holds
        # the minimum element
        min_idx = i
        
        # Iterate through the unsorted portion
        # to find the actual minimum
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
              
                # Update min_idx if a smaller element is found
                min_idx = j
        
        # Move minimum element to its
        # correct position
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

def print_array(arr):
    for val in arr:
        print(val, end=" ")
    print()

if __name__ == "__main__":
    arr = [64, 25, 12, 22, 11]
    
    print("Original array: ", end="")
    print_array(arr)
    
    selection_sort(arr)
    
    print("Sorted array: ", end="")
    print_array(arr)
```

## 3. INSERTION SORT

Algoritma ini selalu mempertahankan sebuah sorted sublist pada posisi bagian bawah dari list. Setiap item baru kemudian akan “di-insert” kembali ke dalam sorted sublist sebelumnya sehingga sorted sublist tersebut menjadi lebih besar satu item.

### Langkah-langkahnya:

1. Mulai dari elemen kedua karena elemen pertama dianggap sudah sorted.<br>
2. Bandingkan elemen kedua dengan elemen pertama, jika elemen kedua lebih kecil maka tukar posisinya.<br>
3. Pindah ke elemen ketiga, bandingkan dengan dua elemen sebelumnya, lalu letakkan pada posisi yang benar.<br>
4. Ulangi proses ini sampai seluruh array menjadi sorted.

```python
def insertShort(data):
    for i in range(1, len(data)): #dimulai dari 1
        print(f"iterasi ke - {i}")
        key = data[i]
        idx = i - 1
        while idx >= 0 and key <= data[idx]:
            data[idx+1] = data[idx]
            idx -= 1 #increment
        data[idx+1] = key

        print(data)
        print()
    return data

data = [10, 4, 5, 9, 1, 8, 2]
sorted_data = insertShort(data)
print(sorted_data)
```
