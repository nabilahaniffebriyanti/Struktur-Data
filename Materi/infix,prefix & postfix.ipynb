# INFIX, PREFIX, DAN POSTFIX

### INFIX
Infix Adalah notasi yang membentuk atas operator dengan operand,dimana operator berada diantara operand.<br>
Contoh :<br>
A-BxC+DxE/F

### PREFIX<b>
Prefix Adalah notasi yang terbentuk atas operator dengan operand, dimana operator didepan operand.<br>
Contoh : <br>

INFIX : A-BxC+DxE/F<br>

A-(BxC)+(DxE)/F<br>
A-(xBC)+(xDE)/F<br>
A-xBC+/xDEF<br>
+-AxBC/xDEF<br>

### POSTFIX
Postfix Adalah notasi yang membentuk atas operator dengan operand, dimana operator berada dibelakang operand.<br>
Contoh : <br>

INFIX : A-BxC+DxE/F<br>

A-(BCx)+(DEx)/F<br>
A-(BCx)+(DEx)/F<br>
ABCx-+DExF/<br>
ABCx-DExF/+<br>

#### Konversi infix ke postfix dengan stack<br>

1. jika token yang dibaca adalah operand maka masukkan operand tersebut ke dalam output string<br>
2. jika token yang dibaca adalah kurung buka maka push kurung buka tersebut ke dalam stack<br>
3. jika token yang dibaca adalah kurang tutup maka pop stack semua token sampai ditemukan kurung buka<br>
4. jika token yang dibaca adalah operator maka :<br>
   a. pop operator-operator yang memiliki precedence lebih tinggi atau sama dan masukkan operator tersebut ke dalam output string<br>
   b. push token operator ke dalam stack<br>
Jika masih terdapat operator pada stack, maka pop operator yang tersisa dan letakkan pada output string<br>

````python
def prec(c):
    if c == '^':
        return 3
    elif c == '/' or c == '*':
        return 2
    elif c == '+' or c == '-':
        return 1
    else:
        return -1

def isRightAssociative(c):
    return c == '^'

def infixToPostfix(s):
    st = []
    res = []

    for c in s:
        # If operand, add to result
        if ('a' <= c <= 'z') or ('A' <= c <= 'Z') or ('0' <= c <= '9'):
            res.append(c)

        # If '(', push to stack
        elif c == '(':
            st.append('(')

        # If ')', pop until '('
        elif c == ')':
            while st and st[-1] != '(':
                res.append(st.pop())
            st.pop()

        # If operator
        else:
            while st and st[-1] != '(' and \
                (prec(st[-1]) > prec(c) or (prec(st[-1]) == prec(c) \
                                    and not isRightAssociative(c))):
                res.append(st.pop())
            st.append(c)

    # Pop remaining operators
    while st:
        res.append(st.pop())

    return ''.join(res)

if __name__ == '__main__':
    exp = "a*(b+c)/d"
    print(infixToPostfix(exp))
````

````python
#IMPLEMENTASI KE2
from pythonds.basic.stack import Stack

def infixToPostfix(infixexpr):
    prec = {}
    prec["*"] = 3
    prec["/"] = 3
    prec["+"] = 2
    prec["-"] = 2
    prec["("] = 1

    opStack = Stack()
    postfixList = []
    tokenList = infixexpr.split()

    for token in tokenList:
        if token in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" or token in "0123456789":
            postfixList.append(token)

        elif token == '(':
            opStack.push(token)

        elif token == ')':
            topToken = opStack.pop()
            while topToken != '(':
                postfixList.append(topToken)
                topToken = opStack.pop()

        else:
            while (not opStack.isEmpty()) and \
                  (prec[opStack.peek()] >= prec[token]):
                postfixList.append(opStack.pop())
            opStack.push(token)

    while not opStack.isEmpty():
        postfixList.append(opStack.pop())

    return " ".join(postfixList)


def postfixEval(postfixExpr):
    operandStack = Stack()
    tokenList = postfixExpr.split()

    for token in tokenList:
        if token in "0123456789":
            operandStack.push(int(token))
        else:
            operand2 = operandStack.pop()
            operand1 = operandStack.pop()
            result = doMath(token, operand1, operand2)
            operandStack.push(result)

    return operandStack.pop()


def doMath(op, op1, op2):
    if op == "*":
        return op1 * op2
    elif op == "/":
        return op1 / op2
    elif op == "+":
        return op1 + op2
    else:
        return op1 - op2


# Test infix to postfix
print(infixToPostfix("A * B + C * D"))
print(infixToPostfix("( A + B ) * C - ( D - E ) * ( F + G )"))

# Test postfix evaluation
print(postfixEval("7 8 + 3 2 + /"))
````

### INFIX TO PREFIX

Prosesnya: lakukan **reverse** terlebih dahulu (dibaca dari belakang).<br>
Kemudian, setelah di-reverse, ubah menjadi **postfix**.<br>
Setelah menjadi postfix, lakukan **reverse** lagi.<br>

### POSTFIX TO INFIX <br>

1. Mulai
2. Siapkan stack kosong<br>
3. Baca ekspresi postfix dari kiri ke kanan.<br>
4. Jika karakter adalah operand (huruf/angka), push ke stack.<br>
5. Jika karakter adalah operator:<br>
Pop operand2 dari stack.<br>
Pop operand1 dari stack.<br>
Gabungkan menjadi (operand1 operator operand2).<br>
Push hasil gabungan kembali ke stack.<br>
6. Ulangi langkah 3–5 sampai semua karakter selesai dibaca.<br>
7. Elemen terakhir di stack adalah hasil ekspresi infix.<br>

````python
#IMPLEMENTASI
def postfix_to_infix(expression):
    stack = []

    for char in expression:
        # jika operand, langsung push ke stack
        if char.isalnum():
            stack.append(char)
        else:
            # jika operator, ambil dua operand dari stack
            op2 = stack.pop()
            op1 = stack.pop()

            # gabungkan menjadi infix
            new_expr = "(" + op1 + char + op2 + ")"

            # push kembali ke stack
            stack.append(new_expr)

    return stack[-1]


# Contoh penggunaan
postfix = "AB*C+"
infix = postfix_to_infix(postfix)

print("Postfix :", postfix)
print("Infix   :", infix)
````

### PREFIX TO INFIX

1. Baca ekspresi **Prefix** dalam urutan terbalik (dari kanan ke kiri).
2. Jika simbol adalah **operand**, maka **push** (masukkan) ke dalam **stack**.
3. Jika simbol adalah **operator**, maka **pop** (ambil) dua operand dari **stack**.
4. Buat sebuah string dengan menggabungkan kedua operand dan operator di antara keduanya.

   ```
   string = (operand1 + operator + operand2)
   ```
5. Masukkan kembali (**push**) string hasil tersebut ke dalam **stack**.
6. Ulangi langkah-langkah di atas sampai akhir ekspresi **Prefix**.
7. Pada akhirnya, **stack** hanya akan berisi **1 string**, yaitu hasil ekspresi tersebut.

````python
# Python Program to convert prefix to Infix
def prefixToInfix(prefix):
    stack = []
    
    # read prefix in reverse order
    i = len(prefix) - 1
    while i >= 0:
        if not isOperator(prefix[i]):
            
            # symbol is operand
            stack.append(prefix[i])
            i -= 1
        else:
          
            # symbol is operator
            str = "(" + stack.pop() + prefix[i] + stack.pop() + ")"
            stack.append(str)
            i -= 1
    
    return stack.pop()

def isOperator(c):
    if c == "*" or c == "+" or c == "-" or c == "/" or c == "^" or c == "(" or c == ")":
        return True
    else:
        return False

# Driver code
if __name__=="__main__":
    str = "*-A/BC-/AKL"
    print(prefixToInfix(str))
    
# This code is contributed by avishekarora
````

### EVALUASI PREFIX

Langkah 1: Letakkan penunjuk (pointer) P di bagian paling akhir.

Langkah 2: Jika karakter pada P adalah sebuah operand, masukkan (push) ke dalam Stack.

Langkah 3: Jika karakter pada P adalah sebuah operator, keluarkan (pop) dua elemen dari Stack. Operasikan kedua elemen tersebut sesuai dengan operatornya, lalu masukkan (push) hasilnya kembali ke dalam Stack.

Langkah 4: Kurangi P sebanyak 1 (geser ke kiri) dan ulangi ke Langkah 2 selama masih ada karakter yang tersisa untuk dipindai dalam ekspresi tersebut.

Langkah 5: Hasil akhirnya tersimpan di bagian atas (top) Stack, kembalikan nilai tersebut.

Langkah 6: Selesai.

````python
#IMPLEMENTASI
def evaluasi_prefix(expr):
    stack = []

    # membaca ekspresi dari kanan ke kiri
    for char in reversed(expr):
        
        # jika operand (angka)
        if char.isdigit():
            stack.append(int(char))

        # jika operator
        else:
            op1 = stack.pop()
            op2 = stack.pop()

            if char == '+':
                hasil = op1 + op2
            elif char == '-':
                hasil = op1 - op2
            elif char == '*':
                hasil = op1 * op2
            elif char == '/':
                hasil = op1 / op2

            stack.append(hasil)

    return stack.pop()


# Contoh penggunaan
prefix = "-+7*45+20"
hasil = evaluasi_prefix(prefix)

print("Prefix :", prefix)
print("Hasil  :", hasil)
````

### EVALUASI INFIX

1. Mulai.
2. Siapkan dua buah **stack**:

   * **Stack operand** (untuk menyimpan angka).
   * **Stack operator** (untuk menyimpan operator).
3. Baca ekspresi **infix dari kiri ke kanan**.
4. Jika simbol adalah **operand (angka)**, maka **push ke stack operand**.
5. Jika simbol adalah **"("**, maka **push ke stack operator**.
6. Jika simbol adalah **")"**, maka:

   * Pop operator dari stack operator.
   * Pop dua operand dari stack operand.
   * Lakukan operasi sesuai operator.
   * Push hasilnya kembali ke stack operand.
   * Ulangi sampai menemukan **"("**.
7. Jika simbol adalah **operator (+, -, *, /)**:

   * Selama prioritas operator di stack **lebih tinggi atau sama**, lakukan operasi terlebih dahulu.
   * Pop operator dan dua operand, hitung hasilnya, lalu push kembali ke stack operand.
   * Setelah itu **push operator baru ke stack operator**.
8. Ulangi langkah di atas sampai seluruh ekspresi selesai dibaca.
9. Jika masih ada operator di stack operator:

   * Pop operator dan dua operand.
   * Hitung hasilnya lalu push kembali ke stack operand.
10. Nilai terakhir pada **stack operand** adalah **hasil evaluasi infix**.
11. Selesai.

````python
#IMPLEMENTASI
# fungsi untuk menentukan prioritas operator
def precedence(op):
    if op in ('+', '-'):
        return 1
    if op in ('*', '/'):
        return 2
    return 0

# fungsi untuk melakukan operasi
def operasi(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        return a / b

def evaluasi_infix(expr):
    operand = []
    operator = []

    i = 0
    while i < len(expr):
        char = expr[i]

        # jika angka
        if char.isdigit():
            operand.append(int(char))

        # jika tanda kurung buka
        elif char == '(':
            operator.append(char)

        # jika tanda kurung tutup
        elif char == ')':
            while operator[-1] != '(':
                op = operator.pop()
                b = operand.pop()
                a = operand.pop()
                operand.append(operasi(a, b, op))
            operator.pop()

        # jika operator
        else:
            while (operator and precedence(char) <= precedence(operator[-1])):
                op = operator.pop()
                b = operand.pop()
                a = operand.pop()
                operand.append(operasi(a, b, op))
            operator.append(char)

        i += 1

    # proses sisa operator
    while operator:
        op = operator.pop()
        b = operand.pop()
        a = operand.pop()
        operand.append(operasi(a, b, op))

    return operand.pop()


# contoh penggunaan
infix = "3+5*2"
hasil = evaluasi_infix(infix)

print("Infix :", infix)
print("Hasil :", hasil)
````
