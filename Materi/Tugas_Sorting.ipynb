# SORTING

### Waktu yang diperlukan untuk mengurutkan 100-500 data dari Bubble Sort, Selection Sort, Insertion Sort, Merge Sort dan Quick Sort untuk urutan ascending

```python
# BUBBLE SORT
def bubble_sort(arr):

    data = arr.copy()
    count = 0

    n = len(data)

    for i in range(n):
        for j in range(0, n - i - 1):

            count += 1

            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]

    return count

# SELECTION SORT
def selection_sort(arr):

    data = arr.copy()
    count = 0

    n = len(data)

    for i in range(n):

        min_index = i

        for j in range(i + 1, n):

            count += 1

            if data[j] < data[min_index]:
                min_index = j

        data[i], data[min_index] = data[min_index], data[i]

    return count

# INSERTION SORT
def insertion_sort(arr):

    data = arr.copy()
    count = 0

    for i in range(1, len(data)):

        key = data[i]
        j = i - 1

        while j >= 0 and key < data[j]:

            count += 1

            data[j + 1] = data[j]
            j -= 1

        data[j + 1] = key

    return count

# MERGE SORT
def merge_sort(arr):

    count = [0]

    def merge(data):

        if len(data) > 1:

            mid = len(data) // 2

            left = data[:mid]
            right = data[mid:]

            merge(left)
            merge(right)

            i = j = k = 0

            while i < len(left) and j < len(right):

                count[0] += 1

                if left[i] < right[j]:
                    data[k] = left[i]
                    i += 1
                else:
                    data[k] = right[j]
                    j += 1

                k += 1

            while i < len(left):
                data[k] = left[i]
                i += 1
                k += 1

            while j < len(right):
                data[k] = right[j]
                j += 1
                k += 1

    merge(arr.copy())

    return count[0]

# QUICK SORT
def quick_sort(arr):

    count = [0]

    def quick(data):

        if len(data) <= 1:
            return data

        pivot = data[len(data) // 2]

        left = []
        middle = []
        right = []

        for x in data:

            count[0] += 1

            if x < pivot:
                left.append(x)
            elif x > pivot:
                right.append(x)
            else:
                middle.append(x)

        return quick(left) + middle + quick(right)

    quick(arr.copy())

    return count[0]

# DATA MANUAL
data100 = list(range(100, 0, -1))
data200 = list(range(200, 0, -1))
data300 = list(range(300, 0, -1))
data400 = list(range(400, 0, -1))
data500 = list(range(500, 0, -1))

datas = [
    ("100", data100),
    ("200", data200),
    ("300", data300),
    ("400", data400),
    ("500", data500)
]

print("=" * 85)
print("PERBANDINGAN SORTING")
print("=" * 85)

print(f"{'Data':<10} {'Bubble':<15} {'Selection':<15} {'Insertion':<15} {'Merge':<15} {'Quick':<15}")

for name, data in datas:

    bubble = bubble_sort(data)
    selection = selection_sort(data)
    insertion = insertion_sort(data)
    merge = merge_sort(data)
    quick = quick_sort(data)

    print(f"{name:<10} {bubble:<15} {selection:<15} {insertion:<15} {merge:<15} {quick:<15}")
```
